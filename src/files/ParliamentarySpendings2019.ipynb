# Análise de gastos parlamentares em 2019

```python
import pandas as pd

df = pd.read_csv("2019.csv") 
df
```

## Qual foi o maior tipo de gasto parlamentar

```python
totalByDesc = df.groupby(['descricao']).sum().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
totalByDesc[['vlrDocumento']]
```

```python
totalByDesc[['vlrDocumento']].to_json(force_ascii=False)
```

## Qual o partido que mais gastou absolutamente

```python
totalByParty = df.groupby(['partido'])['partido', 'vlrDocumento'].sum().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
totalByParty.round(2)
```

```python
totalByParty.to_json(force_ascii=False)
```

## Quais foram os parlamentares que mais gastaram

```python
totalByCM = df.groupby(['parlamentar', 'partido']).sum().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
totalByCM[['vlrDocumento']].head(20)
```

```python
totalByCM[['vlrDocumento']].head(20).to_json(force_ascii=False, orient="table")
```

## Estabelecimento que mais forneceu, por descrição

```python
totalByProvider = df.groupby(['fornecedor', 'descricao']).sum().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
totalByProvider[['vlrDocumento']].round(2)
```

```python
totalByProvider[['vlrDocumento']].head(20).to_json(force_ascii=False, orient="table")
```

## Mês em que mais se gastou

```python
totalByMonth = df.groupby(['mes']).sum().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
totalByMonth[['vlrDocumento']].round(2)
```

## Estados que mais gastaram absolutamente

```python
totalByUF = df.groupby(['uf']).sum().sort_values(by=['vlrDocumento'], ascending=False).round(2)
```

```python
totalByUF[['vlrDocumento']].round(2)
```

```python
totalByUF[['vlrDocumento']].round(2).to_json(force_ascii=False)
```

## Média de gasto por descrição

```python
avgByDesc = df.groupby(['parlamentar','descricao']).sum().groupby(['descricao']).mean().sort_values(by=['vlrDocumento'], ascending=False)
```

```python
avgByDesc[['vlrDocumento']].round(2)
```

```python
avgByDesc[['vlrDocumento']].to_json(force_ascii=False)
```

```python
total = df['vlrDocumento'].sum().round(2)
```
